```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
match_data = pd.read_csv('match/matches.csv')
deliveries_data = pd.read_csv('match/deliveries.csv')
```

```python
match_data.head()
```

```python
deliveries_data.head()
```

```python
score_df = pd.merge(deliveries_data, match_data[['id','season', 'winner', 'result', 'dl_applied', 'team1', 'team2']], left_on='match_id', right_on='id')
score_df.player_dismissed.fillna(0, inplace=True)
score_df['player_dismissed'].loc[score_df['player_dismissed'] != 0] = 1
score_df.head(127)
```

```python
player_data = pd.read_csv('espn/IPL Player Stats - 2016 till 2019.csv')
player_data
```

```python
print(score_df.dtypes)
```

```python
print(player_data.dtypes)
```

```python
player_data['season'] = player_data['Tournament'].str.replace('IPL ','')
player_data.head(10)
```

```python
player_data_cleaned = player_data.drop(['Team','Tournament','Matches','Batting Innings','Not Out','Runds Scored','Highest Score','Balls Faced','100','50','0','4s','6s','Bowling Innings','Overs Bowled','Maidens Bowled','Runs Conceded','Wickets Taken','Best Bowling Figures','4+ Innings Wickets','5+ Innings Wickets','Catches Taken','Stumpings Made','season'], axis=1)
player_data_cleaned.head(20)
```

```python
player_data_cleaned.dtypes
```

```python
player_data_cleaned_1 = player_data_cleaned.copy()
player_data_cleaned_1['Batting Average'].loc[player_data_cleaned_1['Batting Average'] == '-'] = 0
player_data_cleaned_1['Batting Strike Rate'].loc[player_data_cleaned_1['Batting Strike Rate'] == '-'] = 0
player_data_cleaned_1['Bowling Average'].loc[player_data_cleaned_1['Bowling Average'] == '-'] = 0
player_data_cleaned_1['Bowling Economy Rate'].loc[player_data_cleaned_1['Bowling Economy Rate'] == '-'] = 0
player_data_cleaned_1['Bowling Strike Rate'].loc[player_data_cleaned_1['Bowling Strike Rate'] == '-'] = 0
player_data_cleaned_1 = player_data_cleaned_1[['Batting Average','Batting Strike Rate','Bowling Average','Bowling Economy Rate','Bowling Strike Rate']].astype(str).astype(float)
player_data_cleaned_1['Player'] = player_data_cleaned['Player']
player_data_cleaned_1.head(20)
```

```python
player_data_final = player_data_cleaned_1.groupby(['Player'], as_index= False).agg('mean')
player_data_final.head()
```

```python
player_data_final
```

```python
dataset_part_1 = pd.merge(score_df,player_data_final[['Batting Average','Batting Strike Rate']], left_on='batsman',right_on='Player')
dataset_part_1
```

```python
score_df.insert(8,"batsman strike rate","NA")
score_df.insert(9,"batsman average","NA")
score_df.head()
```

```python
for i in range(score_df.shape[0]):
    for j in range(player_data_final.shape[0]):
        if score_df['batsman'][i]==player_data_final['Player'][j]:
            score_df['batsman strike rate'][i]=player_data_final['Batting Strike Rate'][j]
            score_df['batsman average'][i]=player_data_final['Batting Average'][j]
```

```python
score_df.head(130)
```

```python
score_df.insert(12,"bowler strike rate","NA")
score_df.insert(13,"bowler average","NA")
score_df.insert(14,"bowler economy rate","NA")
# score_df.head()
for i in range(score_df.shape[0]):
    for j in range(player_data_final.shape[0]):
        if score_df['bowler'][i]==player_data_final['Player'][j]:
            score_df['bowler strike rate'][i]=player_data_final['Bowling Strike Rate'][j]
            score_df['bowler average'][i]=player_data_final['Bowling Average'][j]
            score_df['bowler economy rate'][i]=player_data_final['Bowling Economy Rate'][j]
score_df.head(130)
```

```python
score_df.columns
```

```python
score_df['batsman strike rate'].dtypes
```

```python
battingStrikeRateMean = player_data_final['Batting Strike Rate'].mean()
score_df['batsman strike rate'].loc[score_df['batsman strike rate'] == 'NA'] = battingStrikeRateMean
```

```python
battingAverageMean = player_data_final['Batting Average'].mean()
score_df['batsman average'].loc[score_df['batsman average'] == 'NA'] = battingAverageMean
bowlingStrikeRateMean = player_data_final['Bowling Strike Rate'].mean()
score_df['bowler strike rate'].loc[score_df['bowler strike rate'] == 'NA'] = bowlingStrikeRateMean
bowlingEconomyRateMean = player_data_final['Bowling Economy Rate'].mean()
score_df['bowler economy rate'].loc[score_df['bowler economy rate'] == 'NA'] = bowlingEconomyRateMean
bowlingAverageMean = player_data_final['Bowling Average'].mean()
score_df['bowler average'].loc[score_df['bowler average'] == 'NA'] = bowlingAverageMean
```
