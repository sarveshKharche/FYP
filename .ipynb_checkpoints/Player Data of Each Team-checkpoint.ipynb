```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
espnIPLDataset = pd.read_csv('espn/IPL Player Stats - 2016 till 2019.csv')
espnIPLDataset.head()
```

```python
espnIPLDataset = espnIPLDataset.drop(['Tournament'],axis=1)
espnIPLDataset.head()
```

```python
espnIPLDataV2 = espnIPLDataset.drop(['Matches','Batting Innings','Not Out','Runds Scored','Highest Score','Balls Faced','100','50','0','4s','6s','Bowling Innings','Overs Bowled','Maidens Bowled','Runs Conceded','Wickets Taken','Best Bowling Figures','4+ Innings Wickets','5+ Innings Wickets','Catches Taken','Stumpings Made'], axis=1)
espnIPLDataV2.head()
```

```python

espnIPLDataV2['Batting Average'].loc[espnIPLDataV2['Batting Average'] == '-'] = 0
espnIPLDataV2['Batting Strike Rate'].loc[espnIPLDataV2['Batting Strike Rate'] == '-'] = 0
espnIPLDataV2['Bowling Average'].loc[espnIPLDataV2['Bowling Average'] == '-'] = 0
espnIPLDataV2['Bowling Economy Rate'].loc[espnIPLDataV2['Bowling Economy Rate'] == '-'] = 0
espnIPLDataV2['Bowling Strike Rate'].loc[espnIPLDataV2['Bowling Strike Rate'] == '-'] = 0
espnIPLDataV2 = espnIPLDataV2[['Batting Average','Batting Strike Rate','Bowling Average','Bowling Economy Rate','Bowling Strike Rate']].astype(str).astype(float)
espnIPLDataV2['Player'] = espnIPLDataset['Player']
espnIPLDataV2['Team'] = espnIPLDataset['Team']
espnIPLDataV2.head()
```

```python
espnIPLDataV2.head()
```

```python
espnIPLDataV2.Team.unique()
```

```python
espnIPLDataV2.replace(to_replace='Rising Pune Supergiants',value='Rising Pune Supergiant',inplace=True)
espnIPLDataV2.drop(espnIPLDataV2[espnIPLDataV2['Team']=='Rising Pune Supergiant'].index, inplace = True)
espnIPLDataV2.drop(espnIPLDataV2[espnIPLDataV2['Team']=='Gujarat Lions'].index, inplace = True)
```

```python
espnIPLDataV2.Team.unique()
```

```python
espnIPLDataV2.replace(to_replace='Delhi Daredevils',value='Delhi Capitals',inplace=True)
```

```python
espnIPLDataV2.Team.unique()
```

```python
espnIPLDataV3 = espnIPLDataV2.groupby(['Player'], as_index= False).agg('mean')
espnIPLDataV3.head()
```

```python
espnIPLDataV2
```

```python
espnIPLDataV3
```

```python
espnIPLDataV2 = espnIPLDataV2.reset_index(drop=True)
```

```python
espnIPLDataV3['Team']=""
for i in range(espnIPLDataV3.shape[0]):
    for j in range(espnIPLDataV2.shape[0]):
        if espnIPLDataV3['Player'][i]==espnIPLDataV2['Player'][j]:
            espnIPLDataV3['Team'][i]=espnIPLDataV2['Team'][j]
espnIPLDataV3.head()
```

```python
espnIPLDataV3.to_csv('PlayerDataCon.csv')
```
