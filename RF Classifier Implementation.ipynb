```python
dataset= dataset.drop(['Unnamed: 0',
       'batsman', 'non_striker',
       'bowler'],axis=1)
```

```python
dataset = dataset.dropna(axis=0,how='any')
```

```python
dataset = dataset.drop('result', axis=1)
```

```python
dataset = dataset.drop(['wide_runs', 'bye_runs', 'legbye_runs',
       'noball_runs', 'penalty_runs', 'batsman_runs', 'extra_runs'],axis=1)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split

X = dataset[['match_id', 'season', 'team1', 'team2', 'winner',
       'inning', 'batting_team', 'bowling_team','over', 'ball','is_super_over',
       'batsman strike rate', 'batsman average',
       'bowler strike rate', 'bowler average',
       'bowler economy rate']]
y = dataset[['total_runs','player_dismissed']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

```python
pred_df = pd.DataFrame(y_pred)
pred_df.head()
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test['total_runs'], pred_df[0]))
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test['player_dismissed'], pred_df[1]))
```

```python
import pandas as pd
feature_imp = pd.Series(clf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```
